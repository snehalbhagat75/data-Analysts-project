```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/harvard_reviews.csv')
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df['published_date'] = pd.to_datetime(df['published_date'])

```

#EDA

# which plateform has most of the published review ?

```python
df['published_platform'].value_counts().plot(kind='bar')
```

# Whats the number of rating has most ?

```python
df['rating'].value_counts().plot(kind='bar')
```

# From which platform most of the ratings are published?

```python
sns.displot(data=df , x='rating' , y='published_platform')
```

# How much people gave votes?

```python
df['helpful_votes'].value_counts().plot(kind='bar')
```

# Which title has most of the rating and from which plateform they were gave?

```python
top_title = df['title'].value_counts().head(15).index.tolist()


filtered = df[df['title'].isin(top_title)]

sns.scatterplot(data=filtered , x='title' , y='rating' , hue='published_platform')
plt.xticks(rotation=90)
plt.show()
```

```python
from wordcloud import WordCloud
```

# which words are repeated maximum time?

```python
text_col = df["text"]

text = ' '.join(text_col)

wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
