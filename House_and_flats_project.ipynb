# Hows sqft_living?

```python
top_sqft_living =df['sqft_living15'].value_counts().head(15).index.tolist()
filtered = df[df['sqft_living15'].isin(top_sqft_living)]
sns.boxplot(data=filtered , x='sqft_living15')
plt.xticks()
plt.show()
```

# relation between sqft_lot and sqft_living?

```python
top_sqft_lot =df['sqft_lot15'].value_counts().head(15).index.tolist()
top_sqft_living =df['sqft_living15'].value_counts().head(15).index.tolist()

filtered = df[df['sqft_lot15'].isin(top_sqft_lot) & df['sqft_living15'].isin(top_sqft_living)]
sns.catplot(data=filtered , x='sqft_lot15' , y='sqft_living15')
plt.xticks(rotation=75)
plt.show()
```

# price of bathrooms and bedrooms?

```python
top_price = df['price'].value_counts().head(15).index.tolist()
filtered = df[df['price'].isin(top_price)]

sns.catplot(data=filtered , x='bedrooms' , y='bathrooms' , hue='price')
plt.xticks(rotation=75)
plt.show()
```

# which year build a maximum buildings ?

```python
top_year = df['yr_built'].value_counts().head(30).index.tolist()
filtered = df[df['yr_built'].isin(top_year)]

sns.histplot(data=filtered , x='yr_built')
plt.xticks(rotation=75)
plt.show()
```

# Do year affect price ?

```python
plt.figure(figsize=(11, 5))
top_year = df['yr_built'].value_counts().head(30).index.tolist()
top_price = df['price'].value_counts().head(15).index.tolist()

filtered = df[df['yr_built'].isin(top_year) & df['price'].isin(top_price)]

sns.barplot(data=filtered , x='yr_built' , y='price')

plt.xticks(rotation=75)
plt.show()
```

# Corretion between dataframe

# how much years later house renoveted?

```python
plt.figure(figsize=(11, 5))
top_year = df['yr_built'].value_counts().head(15).index.tolist()
top_yr_renovet = df['yr_renovated'].value_counts().head(15).index.tolist()

filtered = df[df['yr_built'].isin(top_year) & df['yr_renovated'].isin(top_yr_renovet)]

sns.barplot(data=filtered , x='yr_built' , y='yr_renovated')

plt.xticks(rotation=75)
plt.show()
```

# Do floors affect price ?

```python
# plt.figure(figsize=(11, 5))
top_floor = df['floors'].value_counts().head(15).index.tolist()
top_price = df['price'].value_counts().head(15).index.tolist()

filtered = df[df['price'].isin(top_price) & df['floors'].isin(top_floor)]

sns.scatterplot(data=filtered , x='floors' , y='price')

plt.xticks(rotation=75)
plt.show()
```
